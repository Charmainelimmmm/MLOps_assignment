# Exploratory Data Analysis

```python
import pandas as pd
```

```python
# read the csv file
data = pd.read_csv('01_hdb_resale_transactions.csv')
data
```

```python
# remove the unnecessary columns
data = data[['town', 'flat_type', 'storey_range', 'floor_area_sqm', 'lease_commence_date', 'cbd_dist', 'min_dist_mrt','resale_price' ]]
data
```

```python
# count of towns
town_counts = data.value_counts('town')
town_counts
```

```python
# count of flat types
flat_counts = data.value_counts('flat_type')
flat_counts
```

```python
# create a visualisation of count of flat types
import matplotlib.pyplot as plt

# create a bar chart
plt.bar(flat_counts.index, flat_counts.values)

# add labels and title
plt.xlabel('Flat Type')
plt.ylabel('Count')
plt.title('Bar Chart of Flat Types')

# display the chart
plt.show()
```

```python
# count of storey range
data.value_counts('storey_range')
```

```python
# count of lease commence date
data.value_counts('lease_commence_date')
```

```python
# create a histogram for count of lease commence date
plt.hist(data['lease_commence_date'], bins=10, edgecolor='black')

# add labels and title
plt.xlabel('Years')
plt.ylabel('Frequency')
plt.title('Histogram of Years')

# display the histogram
plt.show()
```

# Preprocesssing

```python
from pycaret.regression import *

s = setup(data, target = 'resale_price', 
           session_id = 123,
           bin_numeric_features= ['floor_area_sqm', 'lease_commence_date', 'cbd_dist', 'min_dist_mrt'])
```

# Creating the model

```python
lr = create_model('lr')
```

```python
plot_model(lr)
```

```python
save_model(lr, 'deployment_lin')
```
